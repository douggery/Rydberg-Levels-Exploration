```python
#Cs D2 Transition 6S_{1/2}-->6P_{3/2}
n1=6
l1=0
j1=0.5
mj1=0.5
n2=6
l2=1
j2=1.5
mj2=1.5
q=+1
print("Cs D2 Transition 6S_{1/2}-->6P_{3/2}")
print("====================================")
#Radial Matrix element R_{nlj\rightarrown'l'j'}
print("R_{nlj-n'l'j'} = %.3f ea_0" % atom.getRadialMatrixElement(n1,l1,j1,n2,l2,j2))
#Reduced Matrix Element <l||er||l'>
print("<l||er||l'> = = %.3f ea_0" % atom.getReducedMatrixElementL(n1,l1,j1,n2,l2,j2))
#Reduced Matrix Element <j||er||j'>
print("<j||er||j'> = %.3f ea_0" % atom.getReducedMatrixElementJ(n1,l1,j1,n2,l2,j2))
#Angular Coupling
print("<nljmj|er|n'l'j'mj'> = %.3f ea_0\n" %\
        atom.getDipoleMatrixElement(n1,l1,j1,mj1,n2,l2,j2,mj2,q))

#Cs Rydberg Transition 60S_{1/2}-->60P_{3/2}
n1=60
l1=0
j1=0.5
mj1=0.5
n2=60
l2=1
j2=1.5
mj2=1.5
q=+1

print("Cs Rydberg Transition 60S_{1/2}-->60P_{3/2}")
print("===========================================")
#Radial Matrix element R_{nlj\rightarrown'l'j'}
print("R_{nlj-n'l'j'} = %.3f ea_0" % atom.getRadialMatrixElement(n1,l1,j1,n2,l2,j2))
#Reduced Matrix Element <l||er||l'>
print("<l||er||l'> = = %.3f ea_0" % atom.getReducedMatrixElementL(n1,l1,j1,n2,l2,j2))
#Reduced Matrix Element <j||er||j'>
print("<j||er||j'> = %.3f ea_0" % atom.getReducedMatrixElementJ(n1,l1,j1,n2,l2,j2))
#Angular Coupling
print("<nljmj|er|n'l'j'mj'> = %.3f ea_0\n" % \
        atom.getDipoleMatrixElement(n1,l1,j1,mj1,n2,l2,j2,mj2,q))
```

```python
n_start=45

x=np.linspace(n_start,90)
d=[abs(atom2.getDipoleMatrixElement(n1,l1,j1,mj1,val,l2,j2,mj2,q)) for val in x]
plt.plot(x,d)
plt.title('Transition Matrix Elements for Rb Coupling Transition 5P_{3/2}-->nD_{5/2}')
plt.ylabel('Dipole Element d')
plt.xlabel('Coupling Quantum Number n')


```

```python
x=np.linspace(n_start,90)
L2=atom2.getTransitionWavelength(95, 2, 5/2, 96, 1, 3/2)
f=10**-9*3*10**8/(L2)
print(f)
L2=atom2.getTransitionWavelength(80, 2, 5/2, 96, 1, 3/2)
f=10**-9*3*10**8/(L2)
print(f)
L=[-10**-9*3*10**8/(atom2.getTransitionWavelength(val, 2, 5/2, val+1, 1, 3/2)) for val in x]
plt.plot(x,L)
plt.title('Energy Separation for Rb Rydberg Transition nD_{5/2}-->nP_{3/2}')
plt.ylabel('Frequency (GHz)')
plt.xlabel('Rydberg Quantum Number n')
```
